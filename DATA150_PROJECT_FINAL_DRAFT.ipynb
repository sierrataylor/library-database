Description of Project:
Our goal is to create a database to simulate how a library keeps track of the books in its system. The database has information on the books found in the library and tracks how the copies of these books can move between branches and customers who check them out. In addition to storing information about the books, the database also contains information for the people who interact with the library. It stores information for customers, employees, and the different branches of the library.

```python
%%sql
-- Display the genre of each book. 
SELECT Books.title, Genre.name
FROM Books JOIN Genre ON Books.genre_id = Genre.genre_id
;
```

```python
%%sql
-- How many books are in each genre?
SELECT Genre.name, count(distinct(title)) as "books_per_genre"
FROM Books JOIN Genre ON Books.genre_id = Genre.genre_id 
GROUP BY Genre.name
;
```

```python
%%sql
-- How many copies of A Midsummer Night's Dream are at the Parma Snow branch?
SELECT count(isbn) as 'copies_here'
FROM Copies
WHERE branch = 3 and isbn in
    (
    SELECT isbn 
    FROM Books
    WHERE title = "A Midsummer Night's Dream"
    )
;
```

```python
%%sql
-- Which books have not been checked out?
SELECT title, isbn
FROM Books
WHERE isbn in 
    (
        SELECT isbn
        FROM Copies
        WHERE isOut == 0
    )
;
```

```python
%%sql
-- What is the name and card number for customers at the Parma branch?
SELECT name, card_number
FROM Customer
WHERE card_number in 
    (
        SELECT customer_number
        FROM Checkout
        WHERE at_branch = 5 
    )
;
```

```python
%%sql
SELECT DISTINCT isbn, count(isbn) as 'num_copies', branch
FROM COPIES
GROUP BY isbn, branch
;
```

```python
%%sql
-- What books did Mira Cortes check out?
SELECT title 
FROM Books
WHERE isbn in
    (
        SELECT isbn
        FROM Checkout
        WHERE customer_number in
            (
                SELECT card_number
                FROM Customer
                WHERE name = "Mira Cortes"
            )
    )
;
```

```python
%%sql 
-- What are the 5 oldest books in the library's system?
SELECT title, year_published 
FROM Books
ORDER BY year_published 
LIMIT 5
;
```

```python
%%sql
-- Which branches have an employee who is a librarian?
SELECT branch.name
FROM branch, employees
WHERE employees.branch = branch.branch_id AND employees.position == 'Librarian'
;
```

```python
%%sql
--Who are the employees that work at the Mayfield branch?
SELECT name
FROM Employees
WHERE branch in
    (
        SELECT branch_id
        FROM branch
        WHERE name = "Mayfield Branch"
    )
;
```

```python
%%sql
--Which books are checked out at the Richmond Heights Branch?
SELECT title
FROM Books
WHERE isbn in 
  (
      SELECT isbn
      FROM Copies
      WHERE isOut == 1 and branch in 
      (
          SELECT branch_id
          FROM Branch
          WHERE name = "Richmond Heights Branch"
      )
  )
;
```
